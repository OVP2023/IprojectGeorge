```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons, make_circles, make_classification, make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

#  1

```python
X, y = make_moons(n_samples=2000, noise=0.3, random_state=0)
plt.plot(X[y==0,0],X[y==0,1],'sr')
plt.plot(X[y==1,0],X[y==1,1],'sb')
plt.show()
```

```python
#K_neighbors
X_train, X_test, y_train, y_test = train_test_split( X, y,random_state=1, test_size=0.4)
cls=KNeighborsClassifier(n_neighbors=5)
cls.fit(X_train, y_train)
y_pred=cls.predict(X_test)
score = cls.score(X_test, y_test)
print(score)
plt.plot(X_test[y_test==0,0],X_test[y_test==0,1],'sr')
plt.plot(X_test[y_test==1,0],X_test[y_test==1,1],'sb')
plt.plot(X_test[y_pred==0,0],X_test[y_pred==0,1],'.r')
plt.plot(X_test[y_pred==1,0],X_test[y_pred==1,1],'.b')
plt.show()

```

```python
#K_neighbors
# test of k
k_max=50
J_train=[]
J_test=[]
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.4, random_state=100)
for i in range(1,k_max):
  cls=KNeighborsClassifier(n_neighbors=i)
  cls.fit(X_train, y_train)
  y_pred_train=cls.predict(X_train)
  y_pred_test=cls.predict(X_test)
  J_train.append(1-cls.score(X_train, y_train))
  J_test.append(1-cls.score(X_test, y_test))
plt.plot(J_train,'m')
plt.plot(J_test,'g')
plt.show()
print(J_test)
```

```python
#Desicion tree
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.4, random_state=100)
cls=DecisionTreeClassifier(max_depth=10)
cls.fit(X_train, y_train)
y_pred=cls.predict(X_test)
score = cls.score(X_test, y_test)
print(score)
plt.plot(X_test[y_test==0,0],X_test[y_test==0,1],'sr')
plt.plot(X_test[y_test==1,0],X_test[y_test==1,1],'sb')
plt.plot(X_test[y_pred==0,0],X_test[y_pred==0,1],'.r')
plt.plot(X_test[y_pred==1,0],X_test[y_pred==1,1],'.b')
plt.show()
```

```python
#Desicion tree
# test of max_depth
k_max=30
J_train=[]
J_test=[]
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.4, random_state=10)
for i in range(1,k_max):
  cls=DecisionTreeClassifier(max_depth=i)
  cls.fit(X_train, y_train)
  y_pred_train=cls.predict(X_train)
  y_pred_test=cls.predict(X_test)
  J_train.append(1-cls.score(X_train, y_train))
  J_test.append(1-cls.score(X_test, y_test))
plt.plot(J_train,'m')
plt.plot(J_test,'g')
plt.show()
```

```python

X4, y4 = make_blobs(n_samples=500, centers=[[4, 4], [-4, -4]],cluster_std=2.5)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X4= np.dot(X4, transformation)

plt.plot(X4[y4==0,0],X4[y4==0,1],'sr')
plt.plot(X4[y4==1,0],X4[y4==1,1],'sb')
plt.show()
```

```python
# K_neighbors
k_max=50
min2=1  #1   score = 0,    
min3=1
min4=1
# dataset 2  train  test
X2_train, X2_test, y2_train, y2_test = train_test_split( X2, y2, test_size=0.4, random_state=100)
# dataset 3  train  test
X3_train, X3_test, y3_train, y3_test = train_test_split( X3, y3, test_size=0.4, random_state=100)
# dataset 4  train  test
X4_train, X4_test, y4_train, y4_test = train_test_split( X4, y4, test_size=0.4, random_state=100)
#  n_neighbors   score      KNeighborsClassifier
for i in range(1,k_max):
    #  K_neighbors     2,3,4
    model_KNC2=KNeighborsClassifier(n_neighbors=i)
    model_KNC3=KNeighborsClassifier(n_neighbors=i)
    model_KNC4=KNeighborsClassifier(n_neighbors=i)
    #   train
    model_KNC2.fit(X2_train, y2_train)
    model_KNC3.fit(X3_train, y3_train)
    model_KNC4.fit(X4_train, y4_train)
    #   test
    y2_pred_test=model_KNC2.predict(X2_test)
    y3_pred_test=model_KNC3.predict(X3_test)
    y4_pred_test=model_KNC4.predict(X4_test)
    # n_neighbors      f=(1-model.score(X_test, y_test)),     score   f
    f2=(1-model_KNC2.score(X2_test, y2_test))
    if f2<min2:
        min2=f2
        n_neighbors2=i
    f3=(1-model_KNC3.score(X3_test, y3_test))
    if f3<min3:
        min3=f3
        n_neighbors3=i
    f4=(1-model_KNC4.score(X4_test, y4_test))
    if f4<min4:
        min4=f4
        n_neighbors4=i
```

```python
# 

# score   n_neighbors  
score2=str(model_KNC2.score(X2_test, y2_test))
score3=str(model_KNC3.score(X3_test, y3_test))
score4=str(model_KNC4.score(X4_test, y4_test))
s_n_neighbors2=str(n_neighbors2)
s_n_neighbors3=str(n_neighbors3)
s_n_neighbors4=str(n_neighbors4)

fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(18,6))
ax1.set_title('K_neighbors,DS2,n_neighbors='+s_n_neighbors2+', score='+score2)
ax1.plot(X2_test[y2_test==0,0],X2_test[y2_test==0,1],'sr')   #   test_true        
ax1.plot(X2_test[y2_test==1,0],X2_test[y2_test==1,1],'sb')
ax1.plot(X2_test[y2_pred_test==0,0],X2_test[y2_pred_test==0,1],'.r')
ax1.plot(X2_test[y2_pred_test==1,0],X2_test[y2_pred_test==1,1],'.b')

ax2.set_title('K_neighbors,DS3,n_neighbors='+s_n_neighbors3+', score='+score3)
ax2.plot(X3_test[y3_test==0,0],X3_test[y3_test==0,1],'sr')
ax2.plot(X3_test[y3_test==1,0],X3_test[y3_test==1,1],'sb')
ax2.plot(X3_test[y3_pred_test==0,0],X3_test[y3_pred_test==0,1],'.r')
ax2.plot(X3_test[y3_pred_test==1,0],X3_test[y3_pred_test==1,1],'.b')

ax3.set_title('K_neighbors,DS4,n_neighbors='+s_n_neighbors4+', score='+score4)
ax3.plot(X4_test[y4_test==0,0],X4_test[y4_test==0,1],'sr')
ax3.plot(X4_test[y4_test==1,0],X4_test[y4_test==1,1],'sb')
ax3.plot(X4_test[y4_pred_test==0,0],X4_test[y4_pred_test==0,1],'.r')
ax3.plot(X4_test[y4_pred_test==1,0],X4_test[y4_pred_test==1,1],'.b')
plt.show()

plt.show()
```
